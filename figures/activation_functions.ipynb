```python
import matplotlib.pyplot as plt
import scienceplots
import numpy as np


plt.style.use(['science', 'std-colors'])

plt.rcParams['figure.figsize'] = (3.54, 3.54/1.5)
# plt.rcParams['figure.dpi'] = 600
```

```python


x = np.linspace(-4, 4, 100)
y = np.maximum(0, x)

plt.figure(figsize=(3.54, 2))
plt.plot(x, y)
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel(r'$\phi(x)=\text{ReLU}(x)$')
plt.xlim(-2.5, 2.5)
plt.ylim(-0.25, 2.5)

plt.savefig("figures/relu.png",bbox_inches='tight', dpi=600)
plt.savefig("figures/relu.pdf",bbox_inches='tight', dpi=600)
plt.show()
```

```python
x = np.linspace(-4, 4, 100)
y = 0.5 * x * (1 + np.tanh(np.sqrt(2/np.pi) * (x + 0.044715 * np.power(x, 3))))
plt.figure(figsize=(3.54, 2))
plt.plot(x, y)
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel(r'$\phi(x)=\text{GELU}(x)$')
plt.xlim(-2.5, 2.5)
plt.ylim(-0.25, 2.5)
plt.savefig("figures/gelu.png",bbox_inches='tight', dpi=600)
plt.savefig("figures/gelu.pdf",bbox_inches='tight', dpi=600)
plt.show()
```
