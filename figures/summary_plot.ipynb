```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import scienceplots

plt.style.use(["science", "std-colors", "grid"])
plt.rcParams["axes.axisbelow"] = True
# plt.rcParams["axes.grid"] = True
# print( plt.rcParams['figure.figsize'])
# plt.rcParams['figure.figsize'] = (3.5, 2.625)
```

```python
data = [
    [   # K78k Test
        0.5603,
        0.5721,
        0.8507,
        0.9426,
        0.9536,
    ], 
    [   # Im2GPS
        0.2025,
        0.2532,
        0.4304,
        0.8608,
        0.8987,
    ],   
    [   # Im2GPS2k
        0.1745,
        0.1970,
        0.3415,
        0.7330,
        0.7470,
    ],  
    ]

bar_width = 0.25
y_pos = np.arange(len(data[0]))

fig, ax = plt.subplots(figsize=(3.5*1.25, 2.625/1.35))

ax.vlines(data[0][-1], -1, 4 + bar_width, colors="#0C5DA5", linestyle=(1, (2,1)), linewidth=1, zorder=-1)
ax.vlines(data[1][-1], -1, 4, colors="#00B945", linestyle=(1, (2,1)), linewidth=1, zorder=-1)
ax.vlines(data[2][-1], -1, 4 - bar_width, colors="#FF9500", linestyle=(1, (2,1)), linewidth=1, zorder=-1)

ax.barh(y_pos + bar_width, data[0], bar_width, label="Kerger Test Set")
ax.barh(y_pos, data[1], bar_width, label=r'\textsc{Im2GPS}')
ax.barh(y_pos - bar_width, data[2], bar_width, label=r'\textsc{Im2GPS2k}')


ax.set_xlabel('Continent-level Accuracy')
ax.set_yticks(y_pos)
ax.set_yticklabels(
    [
    r'\textbf{\textit{Sec. 4.3.}: BaseCNN (Initial Prototype)}',
    r'\textit{Sec. 5.1.}: ViT Architecture',
    r'\textit{Sec. 5.2.}: Pre-Trained Weights \& Model Size',
    r'\textit{Sec. 5.3.}: CLIP, Model Size, Higher Resolution',
    r'\textbf{\textit{Sec. 5.4.}: Additional Training Data}'
    ]
)
ax.set_xlim(0,1)
ax.set_ylim(-0.65, 4.65)

for tick in ax.yaxis.get_majorticklabels():
    tick.set_horizontalalignment("left")

ax.tick_params(axis='y', pad=225)   

plt.legend(loc='upper right', bbox_to_anchor=(1.48, 1.05))

plt.savefig("figures/optimization-summary.png",bbox_inches='tight', dpi=600)
plt.savefig("figures/optimization-summary.pdf",bbox_inches='tight', dpi=600)
plt.show()
```
